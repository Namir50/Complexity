**In computer science, the time complexity is the computational complexity that describes the amount of computer time it takes to run an algorithm**

There are many notations to compute time complexity but most of the time we have Big O

**Types of analysis**


1.   Experimental analysis
2.   Theoratical analysis



Experimental Analysis: The time complexity here may vary from computer to computer

```python
import time
s = time.time()
a = [1,2,3,1,5,6]
duplicate = None
for i in range(len(a)):
  for j in range(i+1,len(a)):
    if a[i] == a[j]:
      duplicate = a[i]
print(duplicate)
e = time.time()
print(e-s)
```

Theoratical Analysis

Important rules:

1.   Drop the constant
2.   Drop non dominant items



```python
a = [1,2,3,1,5,6]              #k      (k means constant or for assignment)
duplicate = None               #k1
for i in range(len(a)):        #n      (we assign n for loops)
  for j in range(i+1,len(a)):  #n      (we do not put n2 since it is nested loops)
    if a[i] == a[j]:           #k2
      duplicate = a[i]         #k4
print(duplicate)
```

k+k1+k2+k3 + n^2  = n^2 = O(n^2)    
k is constants

if it was k+k1+k2+k3 + n^2 + n then the answer would still be n^2 as n^2 is the dominant term

**Big O Notation**

Big O notation is a mathematical notation that describes the limiting behaviour of a function when the argument tends towards a particular value or infinity

1. Constant time: O(1)
2. Linear time: O(n)
3. Logarithmic time: O(log n)
4. Quadratic time: O(n^2)
5. Cubic time: O(n^3)
